```python
# as number of filling values is more we have to handle them -- if missing values are less we can drop that cell 
#handeling age missing data with fillna 
#Dropping cabin column as its very time consuming to impute correct cabin numbers and deck number 
```

```python
df['Age'].fillna(df['Age'].median(),inplace = True)
df['Fare'].fillna(df['Fare'].median(),inplace = True)
df.drop('Cabin',axis = 1,inplace = True)
```

```python
#As the dataset is now clean and Has been Handeled with Missing data using Fillna function
#Start the Exploratory Data Analysis
```

Steps to do in EDA

```python
sns.countplot(x='Survived',hue='Sex',data = df)
plt.title("Survival By Sex")
plt.show()
```

```python
sns.countplot(x='Survived', data = df)
plt.title("Total Survival Counts")
plt.show()
```

```python
sns.countplot(x='Survived',hue=df['Pclass'].astype(str),data = df)
plt.title("Survival By Class")
plt.show()
```

```python
#Correlation Heatmap
numeric = df.select_dtypes(['int64','float64'])
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(numeric,annot = True)
cmap = ("coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
